# Particle Filter
## Task 1
Create the Particle set of size $N=1000$ and with equal weights, given that $x\sim \mathcal{N}(0,1)$.

Plot the histogram of values.

```python
import numpy as np
import matplotlib.pyplot as plt

set_x = np.random.normal(0, 1, 1000)
plt.hist(set_x, bins=30, edgecolor='black')
plt.show()
```

## Task 2
Propagate the particle set, given the following 1d system: $y = (x+\eta)^2$, where $\eta\sim\mathcal{N}(0,2)$ and obtain $\bar{\text{bel}}(y)$

```python
eta_noice = np.random.normal(0, np.sqrt(2), 1000)
y = (set_x + eta_noice)**2
plt.hist(y, bins=30, edgecolor='black')
plt.show()
```

What can you say about this PDF?

## Task 3
Now, let's consider the observation function $h(y) = y$, the probablistic model for the likelihood $p(z|y) = \mathcal{N}(z;h(y),2^2)$ and the observation $z=5$. Calculate the new weighs of the particle set that represent the posterior belief $\text{bel}(y)$

```python
from scipy.stats import norm

obs_func = y
likelihood_dist = norm(loc=5, scale=2)
weights = likelihood_dist.pdf(obs_func)

plt.hist(weights, bins=30, edgecolor='black')
plt.show()
```

```python
print("mean =", y.mean())
print("mean weights =", (weights*y).sum()/weights.sum())
```
